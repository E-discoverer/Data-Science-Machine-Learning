The pandas library provides a fast and powerful tool for a vast variety of data analysis, manipulation, and even data visualization! Yes, pandas enables you to accomplish most of the tasks of the data analysis pipeline:

1. Loading dataset from different sources like a csv file or URL More info.

#### DATA VISUALIZATION IN PYTHON | PANDAS VISUALIZATION | EXAMPLES
In this section, we develop a sample pandas DataFrame to represent our dataset. We will then demonstrate different types of plots for pandas visualization.

##### Importing required libraries.

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np
```

Create a sample **DataFrame** based on random numbers. we use Numpy library to generate random numbers.

```python
numbers = 10
np.random.seed(123)
df = pd.DataFrame(np.random.randn(numbers, 3), index=np.arange(numbers), columns = list('ABC'))

print(f'Our sample DataFrame with {numbers} rows:')
df
```

#### PANDAS VISUALIZATION | LINE PLOT.

```python
df.plot();
```

> When working with *pandas* visualization, note that the *DataFrame plot* method, plots a line-chart by default. 

> You can specify what kind of plot to produce by passing the proper value to the argument kind.The following types of plot is supported by *pandas*:

'line' : line plot (default)

'bar' : vertical bar plot

‘barh' : horizontal bar plot

'hist' : histogram

'box' : boxplot

'kde' : Kernel Density Estimation plot

'density' : same as ‘kde’

'area' : area plot

'pie' : pie plot

'scatter' : scatter plot (DataFrame only)

'hexbin' : hexbin plot (DataFrame only)

#### PANDAS VISUALIZATION | BAR PLOT

```python
df.plot(kind='bar');
```

You can also produce the bar-plot for a given row(s) of dataset:

```python
df.iloc[5].plot(kind='bar')
```

```python
df.iloc[5:9].plot(kind='bar');
```

Alternatively, you could use the bar method for producing a bar plot:

```python
df.plot.bar();
```

 #### PANDAS VISUALIZATION | STACKED BAR PLOT.
 
 You can pass stacked=True for producing a stacked bar plot:

```python
df.plot.bar(stacked=True);
```

```python
df.plot(kind='bar', stacked=True);
```

#### PANDAS VISUALIZATION | HORIZONTAL BAR PLOT
Similarly, you can use the *barh* method, or pass the *kind='barh'* to plot a horizontal bar-chart:

```python
df.plot.barh();
```

```python
df.plot(kind='barh');
```

#### PANDAS VISUALIZATION | STACKED HORIZONTAL BAR PLOT

In a similar fashion, you can produce a stacked horizontal bar-plot:

```python
df.plot.barh(stacked=True);
```

#### PANDAS VISUALIZATION | FORMATTING OF PLOT

The *pandas DataFrame plot* method has numerous keyword arguments to control the formatting of the returned plot, some of which are shown in the following:

```python
df.plot.bar(
figsize=(6,4), # size of figure(w,h) in inches
xlabel = 'Label for x-axis',
ylabel = 'Label for y-axis',
title = 'Title for figure',
fontsize=13,
rot=45, #rotation for ticks
alpha=0.5, #transparency of a graph plot
);
```

#### PANDAS VISUALIZATION | AREA PLOT

You can create an area plot using the method *area* or by passing *kind=area* to the method plot. 
Note that *stacked=True* is an option only when all columns are all positive or negative (not a mixture of both).

```python
df.plot.area(stacked=False);
```

```python
df.plot(kind='area', stacked=False);
```

#### PANDAS VISUALIZATION | SCATTER PLOT

You can draw scatter plot by using the *scatter* method or passing *kind=scatter* to the *plot* method. 

Note that scatter plot needs numeric columns for both x and y axes to be specified by the x and y keywords.

```python
df.plot.scatter(x='A', y='B');
```

```python
df.plot(kind='scatter', x='A', y='C');
```

#### PANDAS VISUALIZATION | FORMATTING OF SCATTER PLOT.

Here is some formatting for the scatter plot:

c: color of each point

s: size of each point

cmap: type of colormap

```python
df.plot.scatter(x='A', y='B', c='C', cmap='viridis', s=100);
```

You could also impose a variable size of points based on their corresponding values in a given column or a function:

```python
df.plot.scatter(x='A', y='B', c='C', cmap='viridis', s=150*df.C.apply(lambda x: x**2));
```

#### PANDAS VISUALIZATION | BOXPLOT.

You can produce a boxplot in *pandas* using the *boxplot* method or the *plot* method and the key='box' argument.

```python
df.boxplot(grid=False);
```

```python
df.plot(kind='box', grid=True);
```

#### PANDAS VISUALIZATION | HISTOGRAMS

Let's create a new dataframe with much more data points suitable for a histogram demonstration:

```python
np.random.seed(123)
df1 = pd.DataFrame(
{
    'a': np.random.randn(1000) +1,
    'b': np.random.randn(1000),
    'c': np.random.randn(1000) -1,
},
columns = ['a','b','c'],
)
```

You can draw histograms using the plot *hist* method or by passing *kind='hist'* to the plot method:

```python
df1.plot(kind='hist', alpha=0.5);
```

```python
df1.plot.hist(alpha=0.2);
```

You can draw a stacked histogram by passing *stacked=True*. The number of bins can also be altered using the *bins* keyword.

```python
df1.plot.hist(stacked=True, bins=20);
```

```python
df1.plot.hist(stacked=False, bins=25);
```

#### PANDAS VISUALIZATION | KDE PLOT

You can produce a density or Kernel Density Estimation plot by passing *plot.kde*,*plot.density* or *kind='kde'* or *kind='density'*:

```python
df1.plot.kde();
#df1.plot(kind='kde');
```

```python
df1.plot.density();
#df1.plot(kind='density');
```

In case a subset of columns needs to be plotted:

```python
df1[['a', 'c']].plot(kind='kde');
```

#### PANDAS VISUALIZATION | SCATTER MATRIX PLOT

The scatter matrix plot is especially helpful, since it enables you to plot a grid of pairwise correlations in dataset. You can produce a scatter plot matrix by using the scatter_matrix method from the *pandas.plotting* module:

```python
from pandas.plotting import scatter_matrix

scatter_matrix(df1, alpha=0.9, figsize=(8,8), s=100, diagonal='kde');
```

```python
#experiment
from pandas.plotting import scatter_matrix

scatter_matrix(df1, alpha=0.9, figsize=(8, 8), s=100, c='c', cmap='viridis', diagonal='kde');
```

#### PANDAS VISUALIZATION | PIE CHART

Let's create another dataframe more suitable for a pie chart:

```python
df2 = pd.DataFrame({'Sales': [70, 30, 55, 20],
                   'Profit': [700, 600, 1550, 100]},
                  index=['Sneakers', 'Watches', 'Perfumes', 'Shirts'])

df2
```

Here is how to plot pie chart using pandas:

```python
df2.plot.pie(y='Profit', figsize=(5, 5));
```

```python
df2.plot.pie(y='Profit', figsize=(5, 5),
            labels=None,
            autopct='%.2f', #percentages
            fontsize=15,
            );
```

Note that *pandas* visualization is built on top of *matplotlib* library. It mainly wraps *matplotlib* methods with some acceptable flexibility of formatting and customization. However, at the time of writing, the placement of legend is not yet flexible through pandas. 
One can still directly use the features that the matplotlib library provides. For example, we can adjust the legend positioning using the *legend* method:

```python
df2.plot.pie(y='Profit', figsize=(6, 6));
plt.legend(loc = 'lower right');
```

You can also make separate subplots for each column:

```python
df2.plot.pie(subplots=True, figsize=(13, 5));
plt.legend(loc = 'lower right');
```

You may not like the format of this plot, mainly due to the overlap of legends with the pie-chart, as well as the overlap of labels "Sales" and "watches".

Again, a detailed customization of plot's format through pandas may be tricky. For such complex plots and/or formatting, you may prefer using matplotlib library directly, since it provides you with all low-level features and options.

### Data Visualization In Python | Visualization With Seaborn.

*Seaborn* is a Python data visualization library built on top of *matplotlib*. That is, *Seaborn* provides a high-level interface enabling you to draw attractive and informative statistical graphics with a relatively small amount of coding.

In the following, we will present some of the data visualization capabilities provided by *seaborn*.

#### VISUALIZATION WITH SEABORN | LINE PLOT.

You can get a report on available example datasets provided by seaborn as shown below:

```python
sns.get_dataset_names()
```

For this example, let's work with the flights dataset from *seaborn*:

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
# checking for missing values in the age column
print(titanic['age'])
print(titanic['age'].isnull())
```

```python
# Total missing values for each feature
print(titanic.isnull().sum())
```

```python
# Replace missing values using mean 
mean = titanic['age'].mean()

titanic['age'].fillna(mean, inplace=True)
```

```python
#Recheck for missing values in age column

print(titanic['age'].isnull().sum())
```

Here is how to produce a line plot in seaborn to show fare versus age for different classes of passengers:

```python
sns.set_theme(style='darkgrid')
sns.lineplot(x='age', y ='fare', hue ='class', data=titanic);
```

You can see how easy it is to produce a high-quality graph with lots of cool pre-defined formatting features.

#### VISUALIZATION WITH SEABORN | GROUPED BAR PLOTS

Let's load the penguins dataset from seaborn:

```python
penguins = sns.load_dataset('penguins')
penguins
```

Here is how to draw grouped bar plot with error bars in seaborn:

```python
splt = sns.catplot(data=penguins, kind='bar', x='species', y='body_mass_g', hue='sex', ci='sd', palette='dark', alpha=.45, height=5)

splt.set_axis_labels('', 'Body mass(g)');
```

#### VISUALIZATION WITH SEABORN | GROUPED BOXPLOTS.

Let's load the tips dataset from seaborn:

```python
tips = sns.load_dataset('tips')
tips
```

Here is how to draw grouped boxplots in seaborn:

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips);
```

#### VISUALIZATION WITH SEABORN | BOXENPLOT

Let's load the diamonds dataset from seaborn:

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

We can get the list of clarity classes from:

```python
clarity_classes = list(diamonds.clarity.unique())
clarity_classes
```

Now, we can use the *seaborn boxenplot* to draw an enhanced box plot for relatively large datasets, like the diamonds dataset:

```python
sns.boxenplot(x='clarity', y='carat', color='b', order=clarity_classes, scale='linear', data=diamonds);
```

### KERNEL DENSITY WITH MARGINAL HISTOGRAMS.

By using seaborn, you can easily draw density plots on side(s) of a main plot. For example, in the following, you see how to draw kernel density plot with marginal histograms in seaborn:

```python
g = sns.JointGrid(data = penguins, x='body_mass_g', y ='bill_depth_mm', space=0.1)
g.plot_joint(sns.kdeplot, fill=True, cmap='hot');
g.plot_marginals(sns.histplot, color='darkred', alpha=0.8, bins=20);
```

### VISUALIZATION WITH SEABORN | ANNOTATED HEATMAPS.

```python
flights_long = sns.load_dataset('flights')
flights = flights_long.pivot('month','year','passengers') #creates a table with these 3 values, pass as the column

#Draw a heatmap with numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, cmap='hot', fmt='d', linewidths=.5, ax=ax);
```
